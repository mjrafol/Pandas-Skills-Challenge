```python
# Import dependencies
import pandas as pd
```

```python
# Import school and student data csv files
school_data_input = ('Resources/schools_complete.csv')
student_data_input = ('Resources/students_complete.csv')
```

```python
# Read the school and student data files and store in Pandas dataframes
school_data = pd.read_csv(school_data_input)
student_data = pd.read_csv(student_data_input)
```

```python
# Display school data file contents
school_data.head()
```

```python
# Display student data file contents
student_data.head()
```

```python
# Display the record count of school data file
school_data.count()
```

```python
# Display the record count of student data file
student_data.count()
```

```python
# Merge data files into a single dataset
data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
data_complete.head()
```

```python
# Display the record count of merged data file
data_complete.count()
```

```python
# Display the data types of columns
data_complete.dtypes
```

```python
# Convert format of both school and student ID to string and calculate the total number of students
data_complete['School ID'] = data_complete['School ID'].astype(str)
data_complete['Student ID'] = data_complete['Student ID'].astype(str)
data_complete.dtypes
```

```python
# Clean column name title casing
data_complete = data_complete.rename(columns={'student_name' : 'Student Name', 'gender' : 'Gender',
                                             'grade' : 'Grade','school_name' : 'School Name',
                                             'reading_score': 'Reading Score', 'math_score' : 'Math Score',
                                             'type' : 'Type','size' : 'Size','budget' : 'Budget'})

data_complete.columns
```

```python
# Clean column name title casing
school_data = school_data.rename(columns={'school_name' : 'School Name', 'type' : 'Type',
                                          'size' : 'Size','budget' : 'Budget'})
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = data_complete['School ID'].nunique()
total_schools
```

```python
# Calculate the total number of students
total_students = data_complete['Student ID'].nunique()
total_students
```

```python
# Calculate the total budget
total_budget = school_data['Budget'].sum()
total_budget
```

```python
# Calculate the average math score
ave_math_grade = data_complete['Math Score'].mean()
ave_math_grade 
```

```python
# Calculate the average reading score
ave_reading_grade = data_complete['Reading Score'].mean()
ave_reading_grade 
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
math_passers = data_complete.loc[data_complete['Math Score'] >= 70,:]
percentage_math_passers = math_passers['Student ID'].count() / total_students * 100
percentage_math_passers
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
reading_passers = data_complete.loc[data_complete['Reading Score'] >= 70,:]
percentage_reading_passers = reading_passers['Student ID'].count() / total_students * 100
percentage_reading_passers
```

```python
# Calculate the percentage of students who passed math and reading (% overall passing)
overall_passers = data_complete.loc[(data_complete['Math Score'] >= 70) & 
                                    (data_complete['Reading Score'] >= 70) ,:]
percentage_overall_passers = overall_passers['Student ID'].count() / total_students * 100
percentage_overall_passers
```

```python
# Create a dataframe to hold the district summary results
district_summary = pd.DataFrame({'Total Schools' : [total_schools], 'Total Students' : total_students,
                                'Total Budget' : total_budget, 'Average Math Score' : ave_math_grade,
                                'Average Reading Score' : ave_reading_grade, '% Passing Math' : percentage_math_passers,
                                '% Passing Reading' : percentage_reading_passers, '% Overall Passing' : percentage_overall_passers})
district_summary
```

```python
# Display the data types of district summary data frame
district_summary.dtypes
```

```python
# Create a clean dataframe to hold the formatted version of district summary results
district_summary_formatted = district_summary
```

```python
# Clean data formatting
district_summary_formatted['Total Students'] = district_summary['Total Students'].map("{:,.0f}".format)
district_summary_formatted['Total Budget'] = district_summary['Total Budget'].map("${:,.2f}".format)
district_summary_formatted['Average Math Score'] = district_summary['Average Math Score'].map("{:,.2f}".format)
district_summary_formatted['Average Reading Score'] = district_summary['Average Reading Score'].map("{:,.2f}".format)
district_summary_formatted['% Passing Math'] = district_summary['% Passing Math'].map("{:,.2f}%".format)
district_summary_formatted['% Passing Reading'] = district_summary['% Passing Reading'].map("{:,.2f}%".format)
district_summary_formatted['% Overall Passing'] = district_summary['% Overall Passing'].map("{:,.2f}%".format)
district_summary_formatted
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Display the columns of the merged data file
data_complete.columns
```

```python
# Display the columns of the school data file
school_data.columns
```

```python
# Group the data by school name
school_group = data_complete.set_index('School Name').groupby(['School Name'])

# Return the school type
school_type = school_data.set_index('School Name')['Type']

# Calculate the total students per school
student_count = school_group['Student ID'].count()

# Calculate the total school budget per school
school_budget = school_data.set_index('School Name')['Budget']

# Calculate the per student budget per school
student_budget = school_data.set_index('School Name')['Budget'] / school_data.set_index('School Name')['Size']

# Calculate the average math score per school
mathgrade = school_group['Math Score'].mean()

# Calculate the average reading score per school
readinggrade = school_group['Reading Score'].mean()

# Calculate the % passing math per school
mathpassers = data_complete.loc[data_complete['Math Score'] >= 70].groupby(['School Name'])
percent_mathpassers = mathpassers['Student ID'].count() / student_count * 100

# Calculate the % passing reading per school
readingpassers = data_complete.loc[data_complete['Reading Score'] >= 70].groupby(['School Name'])
percent_readingpassers = readingpassers['Student ID'].count() / student_count * 100

# Calculate the % overall passing per school
overallpassers = data_complete.loc[(data_complete['Math Score'] >= 70) & 
                                   (data_complete['Reading Score'] >= 70)].groupby('School Name')
percent_overallpassers = overallpassers['Student ID'].count() / student_count * 100
```

```python
# Create a dataframe to hold the school summary results
school_summary = pd.DataFrame({'School Type' : school_type, 
                               'Total Students' : student_count, 
                               'Total School Budget' : school_budget, 
                               'Per Student Budget' : student_budget, 
                               'Average Math Score' : mathgrade,
                               'Average Reading Score' : readinggrade, 
                               '% Passing Math' : percent_mathpassers,
                               '% Passing Reading' : percent_readingpassers, 
                               '% Overall Passing' : percent_overallpassers})
school_summary.head()
```

```python
# Clean data formatting
school_summary_formatted = school_summary.style.format({'Total Students' : '{:,.0f}',
                                                        'Total School Budget' : '${:,.2f}',
                                                        'Per Student Budget' : '${:,.2f}',
                                                        'Average Math Score' : '{:,.2f}',
                                                        'Average Reading Score' : '{:,.2f}',
                                                        '% Passing Math' : '{:,.2f}%',
                                                        '% Passing Reading' : '{:,.2f}%',
                                                        '% Overall Passing' : '{:,.2f}%'})

school_summary_formatted
```

## Scores by School Spending
* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create bins for average spending ranges
bins = [0, 584, 630, 645, 680]

# Define label name for each bin
labels = ['<$585', '$585-630', '$630-645', '$645-680']
```

```python
# Assign records to bins
school_summary['Spending Ranges (Per Student)'] = pd.cut(school_summary['Per Student Budget'], bins, labels = labels, include_lowest = True)
school_summary.head()
```

```python
# Group school spending
school_spending_group = school_summary.groupby(['Spending Ranges (Per Student)'])

# Calculate the average math score per bin
ave_mathscore_range = school_spending_group['Average Math Score'].mean()

# Calculate the average reading score per bin
ave_readingscore_range = school_spending_group['Average Reading Score'].mean()

# Calculate the % of math passers per bin
mathpasser_range = school_spending_group['% Passing Math'].mean()

# Calculate the % of reading passers per bin
readingpasser_range = school_spending_group['% Passing Reading'].mean()

# Calculate the overall passing rate per bin
overallpassingrate = school_spending_group['% Overall Passing'].mean()
```

```python
# Create a data frame for scores by school spending
school_spending = pd.DataFrame({'Average Math Score' : ave_mathscore_range,
                               'Average Reading Score' : ave_readingscore_range,
                               '% Passing Math' : mathpasser_range,
                               '% Passing Reading' : readingpasser_range,
                               '% Overall Passing' : overallpassingrate})

school_spending
```

```python
# Clean data formatting
school_spending_formatted = school_spending.style.format({'Average Math Score' : '{:,.2f}',
                                                  'Average Reading Score' : '{:,.2f}',
                                                  '% Passing Math' : '{:,.2f}%',
                                                  '% Passing Reading' : '{:,.2f}%',
                                                  '% Overall Passing' : '{:,.2f}%'})

school_spending_formatted
```
